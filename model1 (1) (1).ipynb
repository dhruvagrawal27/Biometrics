```python
import os
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.models as models
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import seaborn as sns
from torch.utils.data import Dataset, DataLoader, random_split
from PIL import Image
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
from tqdm import tqdm
import random
from torch.nn import TransformerEncoder, TransformerEncoderLayer

# Set random seeds for reproducibility
torch.manual_seed(42)
random.seed(42)
np.random.seed(42)

# Define dataset paths
FACE_DIR = "/kaggle/input/visa-multimodal/VISA_Face"
IRIS_DIR = "/kaggle/input/visa-multimodal/VISA_Iris/VISA_Iris"

# Define transformations with augmentation for training
face_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

iris_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(5),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])
])

# Define validation/test transformations (no augmentation)
face_val_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

iris_val_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])
])

class VisaDataset(Dataset):
    def __init__(self, face_dir, iris_dir, face_transform=None, iris_transform=None, 
                 mode='train', split_ratio=0.8, create_pairs=True):
        self.face_dir = face_dir
        self.iris_dir = iris_dir
        self.face_transform = face_transform
        self.iris_transform = iris_transform
        self.create_pairs = create_pairs
        
        # Collect all valid subject IDs that have both face and iris data
        self.subject_ids = []
        for person in os.listdir(face_dir):
            if not os.path.isdir(os.path.join(face_dir, person)):
                continue
                
            # Extract subject ID (e.g., S0001)
            subject_id = person[:5]
            iris_path = os.path.join(iris_dir, person[:10])
            
            if os.path.exists(iris_path) and os.path.exists(os.path.join(iris_path, "L")) and os.path.exists(os.path.join(iris_path, "R")):
                # Only add if there are face images and iris images
                face_imgs = [img for img in os.listdir(os.path.join(face_dir, person)) if img.endswith(".jpg")]
                left_iris = [img for img in os.listdir(os.path.join(iris_path, "L")) if img.endswith(".bmp")]
                right_iris = [img for img in os.listdir(os.path.join(iris_path, "R")) if img.endswith(".bmp")]
                
                if face_imgs and (left_iris or right_iris):
                    self.subject_ids.append(subject_id)
        
        # Split subjects into train and validation
        np.random.shuffle(self.subject_ids)
        split_idx = int(len(self.subject_ids) * split_ratio)
        
        if mode == 'train':
            self.subject_ids = self.subject_ids[:split_idx]
        else:  # validation or test
            self.subject_ids = self.subject_ids[split_idx:]
        
        # Create genuine and impostor pairs
        self.pairs = []
        self._create_sample_pairs()
    
    def _create_sample_pairs(self):
        if not self.create_pairs:
            return
            
        # Generate genuine pairs (same person)
        for subject_id in self.subject_ids:
            # Find all face and iris images for this subject
            face_paths = []
            iris_paths = []
            
            # Look for directories that start with the subject ID
            for person in os.listdir(self.face_dir):
                if person.startswith(subject_id) and os.path.isdir(os.path.join(self.face_dir, person)):
                    face_dir_path = os.path.join(self.face_dir, person)
                    face_imgs = [os.path.join(face_dir_path, img) for img in os.listdir(face_dir_path) if img.endswith(".jpg")]
                    face_paths.extend(face_imgs)
                    
                    iris_path = os.path.join(self.iris_dir, person[:10])
                    if os.path.exists(iris_path):
                        if os.path.exists(os.path.join(iris_path, "L")):
                            left_iris = [os.path.join(iris_path, "L", img) for img in os.listdir(os.path.join(iris_path, "L")) if img.endswith(".bmp")]
                            iris_paths.extend(left_iris)
                        if os.path.exists(os.path.join(iris_path, "R")):
                            right_iris = [os.path.join(iris_path, "R", img) for img in os.listdir(os.path.join(iris_path, "R")) if img.endswith(".bmp")]
                            iris_paths.extend(right_iris)
            
            # Create genuine pairs
            if face_paths and iris_paths:
                # Use up to 5 pairs per subject
                num_pairs = min(5, len(face_paths), len(iris_paths))
                for _ in range(num_pairs):
                    face_img = random.choice(face_paths)
                    iris_img = random.choice(iris_paths)
                    self.pairs.append((face_img, iris_img, 1))  # 1 = genuine
        
        # Generate impostor pairs (different people)
        for _ in range(len(self.pairs)):  # Create equal number of impostor pairs
            subject1, subject2 = random.sample(self.subject_ids, 2)  # Pick two different subjects
            
            # Get random face from subject1
            face_paths = []
            for person in os.listdir(self.face_dir):
                if person.startswith(subject1) and os.path.isdir(os.path.join(self.face_dir, person)):
                    face_dir_path = os.path.join(self.face_dir, person)
                    face_imgs = [os.path.join(face_dir_path, img) for img in os.listdir(face_dir_path) if img.endswith(".jpg")]
                    face_paths.extend(face_imgs)
            
            # Get random iris from subject2
            iris_paths = []
            for person in os.listdir(self.face_dir):
                if person.startswith(subject2) and os.path.isdir(os.path.join(self.face_dir, person)):
                    iris_path = os.path.join(self.iris_dir, person[:10])
                    if os.path.exists(iris_path):
                        if os.path.exists(os.path.join(iris_path, "L")):
                            left_iris = [os.path.join(iris_path, "L", img) for img in os.listdir(os.path.join(iris_path, "L")) if img.endswith(".bmp")]
                            iris_paths.extend(left_iris)
                        if os.path.exists(os.path.join(iris_path, "R")):
                            right_iris = [os.path.join(iris_path, "R", img) for img in os.listdir(os.path.join(iris_path, "R")) if img.endswith(".bmp")]
                            iris_paths.extend(right_iris)
            
            if face_paths and iris_paths:
                face_img = random.choice(face_paths)
                iris_img = random.choice(iris_paths)
                self.pairs.append((face_img, iris_img, 0))  # 0 = impostor
        
        # Shuffle the pairs
        random.shuffle(self.pairs)
    
    def __len__(self):
        return len(self.pairs)
    
    def __getitem__(self, idx):
        face_path, iris_path, label = self.pairs[idx]
        face_img = Image.open(face_path).convert("RGB")
        iris_img = Image.open(iris_path).convert("L")
        
        if self.face_transform:
            face_img = self.face_transform(face_img)
        
        if self.iris_transform:
            iris_img = self.iris_transform(iris_img)
        
        return face_img, iris_img, torch.tensor(label, dtype=torch.long)

# Create datasets
train_dataset = VisaDataset(FACE_DIR, IRIS_DIR, face_transform, iris_transform, mode='train')
val_dataset = VisaDataset(FACE_DIR, IRIS_DIR, face_val_transform, iris_val_transform, mode='val')

# Create dataloaders
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False, num_workers=2)

# Define models for feature extraction
class FaceFeatureExtractor(nn.Module):
    def __init__(self, feature_dim=512):
        super(FaceFeatureExtractor, self).__init__()
        self.model = models.resnet50(pretrained=True)
        self.model.fc = nn.Linear(self.model.fc.in_features, feature_dim)
        self.dropout = nn.Dropout(0.5)
    
    def forward(self, x):
        return self.dropout(self.model(x))

class IrisFeatureExtractor(nn.Module):
    def __init__(self, feature_dim=512):
        super(IrisFeatureExtractor, self).__init__()
        self.model = models.resnet34(pretrained=True)
        # Modify first layer to accept 1-channel input
        self.model.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)
        self.model.fc = nn.Linear(self.model.fc.in_features, feature_dim)
        self.dropout = nn.Dropout(0.5)
    
    def forward(self, x):
        return self.dropout(self.model(x))

# Cross-Attention Fusion Module
class CrossAttentionFusion(nn.Module):
    def __init__(self, feature_dim=512, num_heads=8):
        super(CrossAttentionFusion, self).__init__()
        self.feature_dim = feature_dim
        
        # Transformer encoder for cross-modal attention
        encoder_layer = TransformerEncoderLayer(
            d_model=feature_dim, 
            nhead=num_heads, 
            dim_feedforward=2048,
            dropout=0.1
        )
        self.transformer = TransformerEncoder(encoder_layer, num_layers=2)
        
        # Feature reduction
        self.projection = nn.Sequential(
            nn.Linear(feature_dim*2, feature_dim),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(feature_dim, 256)
        )
    
    def forward(self, face_features, iris_features):
        batch_size = face_features.size(0)
        
        # Reshape for transformer: [seq_len, batch_size, feature_dim]
        x = torch.stack([face_features, iris_features], dim=0)
        
        # Apply transformer for cross-attention
        attended = self.transformer(x)
        
        # Extract attended features and concatenate
        attended_face = attended[0]  # [batch_size, feature_dim]
        attended_iris = attended[1]  # [batch_size, feature_dim]
        
        # Concatenate for fusion
        fused = torch.cat([attended_face, attended_iris], dim=1)  # [batch_size, feature_dim*2]
        
        # Project to final feature space
        return self.projection(fused)  # [batch_size, 256]

# Complete multimodal model
class BiometricFusionModel(nn.Module):
    def __init__(self, face_extractor, iris_extractor, fusion_module, num_classes=2):
        super(BiometricFusionModel, self).__init__()
        self.face_extractor = face_extractor
        self.iris_extractor = iris_extractor
        self.fusion_module = fusion_module
        
        # Final classifier
        self.classifier = nn.Sequential(
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, num_classes)
        )
        
        # Face-only and iris-only classifiers for comparison
        self.face_classifier = nn.Linear(512, num_classes)
        self.iris_classifier = nn.Linear(512, num_classes)
    
    def forward(self, face_input, iris_input, mode='fusion'):
        face_features = self.face_extractor(face_input)
        iris_features = self.iris_extractor(iris_input)
        
        if mode == 'face_only':
            return self.face_classifier(face_features)
        elif mode == 'iris_only':
            return self.iris_classifier(iris_features)
        else:  # fusion mode
            fused_features = self.fusion_module(face_features, iris_features)
            return self.classifier(fused_features)

# Initialize model components
face_extractor = FaceFeatureExtractor()
iris_extractor = IrisFeatureExtractor()
fusion_module = CrossAttentionFusion()
model = BiometricFusionModel(face_extractor, iris_extractor, fusion_module)

# Put model on GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

# Loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(), lr=3e-4, weight_decay=1e-5)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=3)

# Train function
def train_epoch(model, dataloader, criterion, optimizer, device, mode='fusion'):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    
    progress_bar = tqdm(dataloader, desc="Training")
    for face_imgs, iris_imgs, labels in progress_bar:
        face_imgs, iris_imgs, labels = face_imgs.to(device), iris_imgs.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(face_imgs, iris_imgs, mode=mode)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        
        progress_bar.set_postfix({'loss': loss.item(), 'acc': 100 * correct / total})
    
    return running_loss / len(dataloader), 100 * correct / total

# Evaluation function
def evaluate(model, dataloader, criterion, device, mode='fusion'):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for face_imgs, iris_imgs, labels in dataloader:
            face_imgs, iris_imgs, labels = face_imgs.to(device), iris_imgs.to(device), labels.to(device)
            
            outputs = model(face_imgs, iris_imgs, mode=mode)
            loss = criterion(outputs, labels)
            
            running_loss += loss.item()
            
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    
    accuracy = 100 * correct / total
    avg_loss = running_loss / len(dataloader)
    
    return avg_loss, accuracy, all_preds, all_labels

# Training loop
def train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs=10, device='cuda'):
    best_val_acc = 0.0
    history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
    
    # Lists to track different modality performance
    face_only_acc = []
    iris_only_acc = []
    fusion_acc = []
    
    for epoch in range(num_epochs):
        print(f"\nEpoch {epoch+1}/{num_epochs}")
        
        # Train with fusion
        train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device, mode='fusion')
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        
        # Evaluate with fusion
        val_loss, val_acc, val_preds, val_labels = evaluate(model, val_loader, criterion, device, mode='fusion')
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        fusion_acc.append(val_acc)
        
        # Evaluate face-only
        _, face_acc, _, _ = evaluate(model, val_loader, criterion, device, mode='face_only')
        face_only_acc.append(face_acc)
        
        # Evaluate iris-only
        _, iris_acc, _, _ = evaluate(model, val_loader, criterion, device, mode='iris_only')
        iris_only_acc.append(iris_acc)
        
        print(f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%")
        print(f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%")
        print(f"Face-only Acc: {face_acc:.2f}%, Iris-only Acc: {iris_acc:.2f}%")
        
        # Update learning rate based on validation accuracy
        scheduler.step(val_acc)
        
        # Save best model
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), 'best_biometric_model.pth')
    
    # Plot training history
    plot_training_history(history)
    
    # Plot comparison between modalities
    plot_modality_comparison(fusion_acc, face_only_acc, iris_only_acc)
    
    return history

# Plotting functions
def plot_training_history(history):
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('Training and Validation Loss')
    
    plt.subplot(1, 2, 2)
    plt.plot(history['train_acc'], label='Train Accuracy')
    plt.plot(history['val_acc'], label='Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy (%)')
    plt.legend()
    plt.title('Training and Validation Accuracy')
    
    plt.tight_layout()
    plt.savefig('training_history.png')
    plt.show()

def plot_modality_comparison(fusion_acc, face_only_acc, iris_only_acc):
    plt.figure(figsize=(10, 6))
    epochs = range(1, len(fusion_acc) + 1)
    
    plt.plot(epochs, fusion_acc, 'o-', color='blue', label='Cross-Attention Fusion')
    plt.plot(epochs, face_only_acc, 's-', color='green', label='Face Only')
    plt.plot(epochs, iris_only_acc, '^-', color='red', label='Iris Only')
    
    plt.xlabel('Epoch')
    plt.ylabel('Validation Accuracy (%)')
    plt.title('Accuracy Comparison: Face vs. Iris vs. Fusion')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    
    plt.tight_layout()
    plt.savefig('modality_comparison.png')
    plt.show()

# Train model
history = train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs=10, device=device)

# Load best model for final evaluation
model.load_state_dict(torch.load('best_biometric_model.pth'))

# Final evaluation
modes = ['fusion', 'face_only', 'iris_only']
mode_names = ['Cross-Attention Fusion', 'Face Only', 'Iris Only']

for mode, name in zip(modes, mode_names):
    val_loss, val_acc, val_preds, val_labels = evaluate(model, val_loader, criterion, device, mode=mode)
    print(f"\n{name} Final Evaluation:")
    print(f"Accuracy: {val_acc:.2f}%")
    
    # Create confusion matrix
    cm = confusion_matrix(val_labels, val_preds)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Impostor', 'Genuine'], yticklabels=['Impostor', 'Genuine'])
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title(f'Confusion Matrix: {name}')
    plt.savefig(f'confusion_matrix_{mode}.png')
    plt.show()
    
    # Print classification report
    print("\nClassification Report:")
    print(classification_report(val_labels, val_preds, target_names=['Impostor', 'Genuine']))
```
